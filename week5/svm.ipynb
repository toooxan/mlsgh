```python
import pandas as pd
import numpy as np
from sklearn import svm

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
recipes = pd.read_csv("C:/Users/al142367/Documents/recipes_muffins_cupcakes.csv")
recipes
```

```python
sns.lmplot(x = 'Sugar', y = 'Butter', data = recipes, hue = 'Type', fit_reg = False)
```

```python
sns.lmplot(x = 'Flour', y = 'Egg', data = recipes, hue = 'Type', fit_reg = False)
```

```python
sugar_butter = recipes[['Sugar', 'Butter']]

type_label = np.where(recipes['Type'] == 'Muffin', 0, 1)
```

```python
model = svm.SVC(kernel = 'linear')

model.fit(sugar_butter, type_label)
```

```python
w = model.coef_[0]
a = -w[0] / w[1]

xx = np.linspace(5, 30)
yy = a * xx - (model.intercept_[0]) / w[1]

sns.lmplot(x = 'Sugar', y = 'Butter', data = recipes, hue = 'Type', fit_reg = False)
plt.plot(xx, yy, color = 'red')
```

```python
def muffin_or_cupcake(butter, sugar):
    if(model.predict([[butter, sugar]]) == 0):
        print("You are looking at a muffin recipe")
    else:
        print("You are looking at a cupcake recipe")
```

```python
sns.lmplot(x = 'Sugar', y = 'Butter', data = recipes, hue = 'Type', fit_reg = False)
plt.plot(xx, yy, color = 'red')
plt.plot(12, 12, 'yo')
```

```python
recipes_scones = pd.read_csv("C:/Users/al142367/Documents/recipes_muffins_cupcakes_scones.csv")
recipes_scones
```

```python
sugar_butter_scones = recipes_scones[['Sugar', 'Butter']]

type_label = recipes_scones['Type'].map({'Muffin': 0, 'Cupcake': 1, 'Scone': 2})

model = svm.SVC(kernel = 'linear', decision_function_shape = 'ovr')

model.fit(sugar_butter_scones, type_label)

sns.lmplot(x = 'Sugar', y = 'Butter', data = recipes_scones, hue = 'Type', fit_reg = False)
```
